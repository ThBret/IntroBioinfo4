## I/ Notions de base

Une variable est créée avec l'opérateur « **assigner** » que l'on écrit sous la forme d'une flèche composée d'un signe moins et d'un tiret :

```
->
```

Ce symbole peut être orienté dans un sens comme dans l'autre du moment que la flèche pointe vers la variable :

````r
# Assignation de variables
x <- 10
5 -> y

x
y
````

En général, la flèche est orientée à gauche, comme ci-dessous :

````r
x <- 10  # variable numérique
y <- "Bonjour"  # chaîne de caractères
z <- TRUE  # valeur logique
````

Comme avec Python, n'importe quelle variable ou ligne de texte peut être imprimée dans la console avec la fonction :


```
print()
```





````r
# Affichage d'une ligne de texte
print("Hello World!")

# Affichage d'une variable
x <- 10
print(x)

# Résultat d'une opération
x <- 10
y <- 5
print(x + y)

# Test logique
print(x == 10)
````

La console R peut être utilisée comme une calculatrice pour performer tout type d'opération arithmétique :

````r
x <- 10
y <- 5

# Opérations arithmétiques
somme <- x + y
différence <- x - y
produit <- x * y
quotient <- x / y
reste <- x %% y  # Modulo (reste de la division)

# Affichage des résultats
print(paste("Somme :", somme))
print(paste("Différence :", différence))
print(paste("Produit :", produit))
print(paste("Quotient :", quotient))
print(paste("Reste :", reste))
````

Autres fonctions mathématiques de base :

````r
# Fonctions mathématiques de base
print(paste("Valeur absolue de -5 :", abs(-5)))
print(paste("Racine carrée de 16 :", sqrt(16)))
print(paste("Exponentielle de 2 :", exp(2)))
print(paste("Logarithme naturel de 10 :", log(10)))
print(paste("Arrondi de 3.14159 :", round(3.14159)))
````

## III/ Structures de données

### a. Vecteur

Un **vecteur** est une collection d'éléments du même type. Avec R, un vecteur est généralement initialisé avec la fonction :

```
c()
```

C'est la méthode la plus basique pour créer un vecteur. On fournit simplement les éléments du vecteur entre parenthèses, séparés par des virgules.

````r
# Créer un vecteur
vec <- c(1, 2, 3, 4, 5)

# Imprimer le vecteur (avec ou sans print())
print(vec)
vec
````

Imprimer la classe d'un vecteur donnera la classe des éléments qu'il contient :

````r
vec1 <- c(1, 2, 3, 4, 5)
vec2 <- c("A", "B", "C", "C")

# Vérifier la classe
print(class(vec1))
print(class(vec2))
````

Il existe d'autres manières d'initialiser des vecteurs avec R. Pour générer une séquence de valeurs **continues**, on peut utiliser :
```
seq()
```

Cette fonction est surtout utilisée pour créer des séquences numériques. Ses paramètres sont :
* **from** : valeur de départ de la séquence.
* **to** : valeur de fin de la séquence.
* **by** : incrément entre chaque valeur.
* **length.out** : nombre d'éléments dans la séquence.

````r
# Exemple 1 : Séquence de nombres de 1 à 5 avec un incrément de 0.8
seq1 <- seq(from = 1, to = 5, by = 0.8)

seq1

# Exemple 2 : Séquence de nombres de 1 à 9 avec une longueur de 4
seq2 <- seq(1, 9, length.out = 4)

seq2
````

Une façon concise de créer une séquence de nombres entiers consécutifs est d'utiliser l'opérateur :
```
:
```
En sachant que le premier et le deuxième nombre seront compris.

````r
# Séquence de nombres entiers consécutifs de 1 à 10
seq3 <- 1:10

seq3
````

Enfin, on peut créer un vecteur de nombres aléatoires sous une distribution uniforme avec la fonction :
```
runif()
```

````r
# Distribution uniforme (0 <= n <= 1)
runif(5)
````

Ou un vecteur de nombres entiers aléatoires avec la fonction :
```
sample()
```

En sachant que la limite inférieure et la limite supérieure sont inclues.

````r
# Nombres entiers aléatoires avec remplacement
sample(1:3, 5, replace = TRUE)

# Nombres entiers aléatoires sans remplacement
sample(1:10, 5, replace = FALSE)
````

## IV / Indexation

Contrairement à beaucoup de langages de programmation comme Python et C, l'inxedation avec R **commence à 1 et non 0**.

On peut extraire des valeurs d'un vecteur, d'une liste ou d'un tableau avec des crochets **[ ]** :

````r
vec <- 1:10

# Accédez aux éléments d'un vecteur en utilisant []
vec[1]  # Premier élément
vec[3:5]  # Du troisième au cinquième élément
vec[c(1, 3)] # Le premier et troisième éléments
````

On peut utiliser des index négatifs pour conserver tous les éléments d'un vecteur sauf ceux spécifiés :

````r
vec[-1] # Tous les éléments sans le premier
vec[c(-1, -3)] # Tous les éléments sauf le premier et le troisième
````

On peut changer la valeur d'un élément en se référant à son index :

````r
# Modifier la valeur du premier élément du vecteur
vec[1] = 0

print(vec)
````

Contrairement à Python, l'alinéa après le début d'une fonction, d'un boucle ou d'une structure conditionnelle n'est pas nécessaire, même s'il est tout de même souvent inclu.

## VII/ Analyses statistiques de base

Dans cette section, nous allons explorer quelques analyses statistiques de base en utilisant R. Nous travaillerons avec un jeu de données intégré à R pour effectuer des statistiques descriptives et des visualisations.

Le jeu de données "iris" comprend 50 échantillons de chacune des trois espèces d'iris (*Iris setosa*, *Iris virginica* et *Iris versicolor*). Quatre caractéristiques ont été mesurées à partir de chaque échantillon : la longueur et largeur des sépales et la longueur et largeur des pétales, en centimètres.

iris.webp

````r
# Chargement des données
data(iris)

# Aperçu des premières lignes du tableau de données
head(iris)
````

### a. Statistiques descriptives

On peut facilement calculer les statistiques descriptives principales pour chaque variable du jeu de données avec la fonction :

```
summary()
```

````r
# Statistiques descriptives de base
summary(iris)
````

### b. Visualisation des données

Créons des graphiques simples pour explorer visuellement les différences entre les différents groupes (espèces) et les potentielles relations entre les variables.

Dans un premier temps, on peut visualiser la distribution des longeurs de pétales à l'aide d'un **histogramme** :

````r
hist(iris$Petal.Length, main = "Distribution de la longueur des sépales", xlab = "Longueur (cm)", ylab = "Fréquence")
````

On observe un grand nombre de valeurs comprises entre 1 et 2 cm et un autre groupe formé autour de 3-6 cm. Pour voir si cette distribution est liée à la présence d'espèces distinctes, on peut générer un **boxplot** (diagramme en boîte) séparant les valeurs en fonction de l'espèce à laquelle elles appartiennent.

````r
# Boxplot de la largeur des pétales pour chaque espèce
boxplot(Petal.Length ~ Species, data = iris,
        xlab = 'Espèces',
        ylab = 'Longueur du pétale (cm)',
        main = "Longueur des pétales pour chacune des trois espèce d'iris",
        col = c('red', 'green', 'blue'))
````

On observe que les différentes longueurs de pétales sont liées à la présence d'espèces d'iris distinctes. Pour voir si cela s'applique aux autres variables, on peut répéter le processus :

````r
# Créer une grille de 2x2
par(mfrow = c(2, 2))

# Créer les 4 boxplots
boxplot(Petal.Width ~ Species, data = iris,
        xlab = 'Espèces',
        ylab = 'Largeur du pétale (cm)',
        main = "Largeur des pétales",
        col = c('red', 'green', 'blue'))

boxplot(Petal.Length ~ Species, data = iris,
        xlab = 'Espèces',
        ylab = 'Longueur du pétale (cm)',
        main = "Longueur des pétales",
        col = c('red', 'green', 'blue'))

boxplot(Sepal.Width ~ Species, data = iris,
        xlab = 'Espèces',
        ylab = 'Largeur du sépale (cm)',
        main = "Largeur des sépales",
        col = c('red', 'green', 'blue'))

boxplot(Sepal.Length ~ Species, data = iris,
        xlab = 'Espèces',
        ylab = 'Longueur du sépale (cm)',
        main = "Longueur des sépales",
        col = c('red', 'green', 'blue'))

# Réinitialiser les paramètres graphiques (optionnel)
par(mfrow = c(1, 1))
````

Il semble que certaines des variables soient potentiellement corrélées. On peut créer un **nuage de points/diagramme de dispersion** pour représenter graphiquement deux variables X et Y et visualiser une potentielle corrélation :

````r
# Diagramme de dispersion entre la longueur et la largeur du sépale
plot(iris$Sepal.Length, iris$Sepal.Width,
     xlab = 'Longueur du sépale',
     ylab = 'Largeur du sépale',
     main = 'Relation entre la longueur et la largeur du sépale chez les iris',
     pch = 19, col = 'blue')
````

À vue d'œil, on observe aucune relation significative entre ces deux variables. On peut tester avec deux autres variables :

````r
# Diagramme de dispersion entre la longueur du pétale et la longueur du sépale
plot(iris$Petal.Length, iris$Sepal.Length,
     xlab = 'Longueur du sépale (cm)',
     ylab = 'Longueur du pétale (cm)',
     main = 'Relation entre la longueur du pétale et du sépale chez les iris',
     pch = 19, col = 'blue')
````

Là on observe une sorte de relation. On peut aussi distinguer les espèces en colorant différemment les points selon l'espèce :

````r
# Diagramme de dispersion entre la longueur du pétale et la longueur du sépale
plot(iris$Petal.Length, iris$Sepal.Length,
     xlab = 'Longueur du sépale (cm)',
     ylab = 'Longueur du pétale (cm)',
     main = 'Relation entre la longueur du pétale et du sépale chez les iris',
     pch = 19, col = factor(iris$Species))
````

### c. Modèle linéaire

Pour examiner la relation entre ces deux variables, on peut construire un **modèle linéaire** qui exprimera leur lien sous la forme d'une équation mathématique. Pour ce faire, on emploie la fonction :
```r
lm()
```

Le modèle suivant est ajusté pour analyser la relation entre la longueur des sépales (Sepal.Length) et la longueur des pétales (Petal.Length) des iris :

````r
# Construction du modèle linéaire
model <- lm(Sepal.Length ~ Petal.Length, data = iris)

# Résumé du modèle
summary(model)
````

Interprétation des résulats :
1. **L'ordonnée à l'origine** (ou intercept) est de 4.30660. Cela signifie que, selon le modèle, lorsque la longueur des pétales est de 0, la longueur estimée des sépales est d'à peu près 4.3 (pas de signification biologique ici car une longueur de pétale de 0 n'est pas réaliste).

2. **Le coefficient directeur** (ou pente) est de 0.40892. Cela indique que, pour chaque augmentation d'une unité de la longueur des pétales, la longueur des sépales augmente en moyenne de 0.40892 cm.

3. **Significativité statistique** : La **valeur-p** ou **p-value** associée au modèle est extrêmement faible (proche de 0), bien inférieure aux seuils standards de significativité (par exemple 0,05 ou 0,01).


→ On en conclue que la relation linéaire entre la longueur du pétale et la longueur du sépale est statistiquement significative.

On peut visualiser le modèle linéaire en ajoutant une **ligne de régression** au diagramme de dispersion généré plus tôt :

````r
# Ajout de la ligne de régression au graphique précédent
plot(iris$Petal.Length, iris$Sepal.Length,
     xlab = 'Longueur du sépale (cm)',
     ylab = 'Longueur du pétale (cm)',
     main = 'Relation entre la longueur du pétale et du sépale chez les iris',
     pch = 19, col = factor(iris$Species))
abline(model, col = 'red', lwd = 2)  # Ligne de régression
````

On peut confirmer la relation linéaire entre les deux variables en calculant le **coefficient de Pearson** :

````r
# Calcul du coefficient de corrélation
cor(iris$Sepal.Length, iris$Petal.Length)
````

Le coefficient de corrélation varie entre -1 et 1. Une valeur proche de 1 indique une corrélation positive forte, tandis qu'une valeur proche de -1 indique une corrélation négative forte. Une valeur proche de 0 suggère une absence de relation linéaire.

On obtient un coefficient particulièrement élevé (proche de 1), indiquant une corrélation positive significative. Lorsque la longueur du pétale augmente, la longueur du sépale augmente aussi, et vice-versa.
